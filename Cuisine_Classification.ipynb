#### IMPORTING THE MODULES

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report
```

```python
df = pd.read_csv("D:\CERTIFICATES\Dataset_cognifyz.csv")
```

```python
df
```

### DESCRIBING THE DATASET

```python
df.info()
```

```python
df.describe()
```

### CHECKING FOR NULL VALUES

```python
df.isnull().sum()
```

### DROPPING THE NULL VALUES

```python
df.dropna()
```

```python
df.isnull().sum()
```

```python
df.fillna(" ", inplace=True)
```

### ENCODING THE CATEGORICAL VARIABLES

```python
categorical_columns = df.select_dtypes(include=['object']).columns
```

```python
label_encoder = LabelEncoder()
df[categorical_columns] = df[categorical_columns].apply(label_encoder.fit_transform)
```

### SPLITTING THE DATASET INTO INDEPENDENT VARIABLE x AND DEPENDENT VARIABLE y

```python
x = df.drop(['Cuisines', 'Encoded_Cuisine'], axis=1)  
y = df['Encoded_Cuisine'] 
```

### SPLITTING THE DATASET FOR TRAINING AND TESTING PURPOSE

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### CREATING THE RANDOM FOREST CLASSIFIER MODEL

```python
classifier = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
classifier.fit(x_train, y_train)
```

### PREDICTING THE VALUES 

```python
y_pred = classifier.predict(x_test)
```

### CHECKING ACCURACY SCORE ,PRECISION SCORE AND RECALL SCORE

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
```

```python
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
```

### CREATING A CLASSIFICATION REPORT

```python
class_report = classification_report(y_test, y_pred)
print("Classification Report:\n", class_report)
```
